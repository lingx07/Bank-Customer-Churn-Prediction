````python
pdf_text = ''
for i, page in enumerate(doc_reader.pages):
    text = page.extract_text()
    if text:
        pdf_text += text
````

````python
print(pdf_text)
````

````python
def format_text(text):
    # Normalize white spaces to a single space
    text = re.sub(r'\s+', ' ', text.strip())
    # Split the text into sentences by detecting periods or question marks followed by spaces
    sentences = re.split(r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?)\s', text)
    # Join each sentence with ' \ ' at the end
    text = ' \\ '.join(sentence for sentence in sentences if sentence) + ' \\'
    # Remove all newline characters
    formatted_text = text.replace('\n', ' ')
    return formatted_text
````

````python
url = "https://www.health.harvard.edu/topics/diet-and-weight-loss"
html = urlopen(url).read()
soup = BeautifulSoup(html, features="html.parser")

# kill all script and style elements
for script in soup(["script", "style"]):
    script.extract()    # rip it out

# get text
web_text = soup.get_text()

print(web_text)
````

````python
formatted_text = format_text(web_text)
print(formatted_text)
````

````python
text_2 = f"""
{formatted_text}
"""
````

````python
prompt = f"""
Perform the following tasks:
1 - Answer the following math question in one sentence:
Question: Find the area bounded by the curves y = x^2 and y = x by integrating with respect to x. \
2 - Summarize the following text_1 delimited by triple in one sentence. \
3 - Answer the following Question about text_1:
Question: What kind of beverages should we avoid and why? \
4 - Summarize the following text_2 delimited by triple in one sentence. \
5 - Answer the following Question about text_2:
Question: What’s a high-fat weight loss diet? \
6 - Summarize the tasks in one sentence.\

Show the output in this format:
1 -
2 -
3 -
4 -
5 -
6 -

Text:
```{text_1}```
```{text_2}```
"""
````

````python
client = OpenAI(
    api_key = OPENAI_API_KEY,
)

response = client.chat.completions.create(
    messages=[
        {
            "role": "user",
            "content": prompt,
        }
    ],
    model="gpt-3.5-turbo",
)
print("Completion for prompt:")
print(response.choices[0].message.content)
````
